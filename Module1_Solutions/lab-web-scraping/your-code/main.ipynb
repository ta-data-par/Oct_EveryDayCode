####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
import pandas as pd
````

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
response = requests.get(url).content
soup = BeautifulSoup(response)
````

````python
tablehead=soup.find("thead")
#classlist = ['tabev6', 'tabev1', 'tabev2', 'tb_region']

cols = [colnames.text for colnames in tablehead.find_all("th")]
cols
#but we actually don't need these specific columns, because we're going to split date and time

cols2 = ["DATE", "TIME", "LATITUDE", "LONGITUDE", "REGION NAME"]
````

````python
tablebod = soup.find("tbody", id="tbody")
tabletime = [times.text.strip().split() for times in soup.select("tbody td.tabev6 a")][:20]
dates = [dates[0] for dates in tabletime]
times = [times[1] for times in tabletime]
times
````

````python
tablereg = [reg.text.strip() for reg in tablebod.find_all("td", class_="tb_region")][:20]
tablereg
````

````python
tablecoords = [coords.text.strip() for coords in tablebod.find_all("td", class_="tabev1")][:40]
tablelats = [tablecoords[i] for i in range(0,len(tablecoords),2)] #getting the latituds 
tablelong = [tablecoords[j] for j in range(1,len(tablecoords),2)]

````

````python
tablecard = [cards.text.strip() for cards in tablebod.find_all("td", class_="tabev2")][:60]
tablecard2 = [i for i in tablecard if i in ["N","S","W","E"]]  #getting rid of magnitude that was also class tabedv2
tablecardlats =[tablecard2[i] for i in range(0,len(tablecard2),2)]
tablecardlong = [tablecard2[i] for i in range(1,len(tablecard2),2)]
````

````python
latitudes = zip(tablelats,tablecardlats) #zipping latitudes and cardinals together
longitudes = zip(tablelong, tablecardlong)
latitudes = [" ".join(list(lats)) for lats in latitudes]
longitudes = [" ".join(list(longs)) for longs in longitudes]
````

````python
dftest = pd.DataFrame({'DATE': dates, 'TIME': times, 'LATITUDE':latitudes, 'LONGITUDE':longitudes, 'REGION':tablereg})
dftest
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url ='https://hackevents.co/search/anything/anywhere/anytime'
````

````python

response = requests.get(url).content
soup = BeautifulSoup(response, "html.parser")
soup
````

````python
title = [i.text for i in soup.select("div.row h3")]
city = [j.text.strip().split(",")[0] for j in soup.select("div.row p:has(i.fa-map-marker-alt)")]
country = [j.text.strip().split(",")[1] for j in soup.select("div.row p:has(i.fa-map-marker-alt)")]
when1 =[j.text.strip().split()[1] for j in soup.select("div.row p:has(i.fa-calendar-check)")]
when2 = [j.text.strip().split()[1] for j in soup.select("div.row p:has(i.fa-calendar-times)")]
when1[0]
````

````python
from datetime import datetime
````

````python
# transforming dates into datetime objects
dtstarts = [datetime.strptime(starts, '%m/%d/%Y')for starts in when1]
dtends = [datetime.strptime(ends, '%m/%d/%Y')for ends in when2]
# calculating time delta and transforming into days
length = [(dtends[i]-dtstarts[i]).days for i in range(len(dtends))]
length
````

````python
df = pd.DataFrame({"TITLE":title, "CITY":city, "COUNTRY":country, "DATE START":when1, "DATE END":when2, "DURATION IN DAYS":length})
df
````
