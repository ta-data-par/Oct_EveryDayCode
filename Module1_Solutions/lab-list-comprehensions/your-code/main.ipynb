### 9. Use a list comprehension to add a new column (20) to the data frame whose values are the values in column 19 minus 0.1. Display the top 10 rows of the resulting data frame.

```python
#CORRECTION
lst=[]
for i, item in data.iterrows():
    lst.append(item[19] - 0.1)
#19th column of every single row, which is i
#accessing column 19 by its index
#also possible to access it by its name, in strg form

    
data['20']=lst
```

```python
#with list comprehension 
#watch out for column numbers, names, etc
#with additionnal index column for example, 
data['20']=[item['19'] - 0.1 for i,item in data.iterrows()]
```

```python
#different solution
#looping over column 19, doesn't need two variables (i and item)
#works with only one column
#you will often need all or several columns in a same row

data['20']=[var-0.1 for var in data['19']]
```

```python
data["20"] = [j['19'] - 0.1 for i,j in data.iterrows()]

print(data.head(10))
```
