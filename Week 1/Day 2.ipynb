```python
lst_lst=[[1,2,3],[1,2,3,4],['yes','no']]
```

```python
lst=[]
for i in lst_lst:
    if len(i)==3:
        for j in i:
            lst.append(j+1)
    elif len(i)==4:
        for j in i:
            lst.append(j**2)
    else:
        for j in i:
            lst.append(j)
```
